a) Consider the following dynamical system 

$$
x_{n+1} = 
\begin{cases} 
    -x_n + 1 + u_n & \text{if } -2 \leq -x_n + 1 + u_n \leq 2 \\
    2 & \text{if } -x_n + 1 + u_n > 2 \\
    -2 & \text{else}
\end{cases}
$$

where $ x_n \in \{-2, -1, 0, 1, 2\} $ and $ u_n \in \{-1, 0, 1\} $, and the cost function

$$
J = \sum_{k=0}^2 \left( 2|x_k| + |u_k| \right) + x_3^2

$$

Use the dynamic programming algorithm to solve the finite horizon optimal control problem that minimizes \( J \). Show the different steps of the algorithm and present the results in a table including the cost-to-go and the optimal control at every stage.



```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Solution:$\\$
Given: $ J = \sum_{k=0}^2 \left( 2|x_k| + |u_k| \right) + x_3^2 $, $ x_n \in \{-2, -1, 0, 1, 2\} $ and $ u_n \in \{-1, 0, 1\} $

```python
u=np.array([-1, 0, 1])
x=np.array([-2,-1,0,1,2])

u2=np.zeros(5)
u1=np.zeros(5)
u0=np.zeros(5)
J2=np.zeros(5)
J1=np.zeros(5)
J0=np.zeros(5)

J3=[]
for i in x:
    i*=i
    J3.append(i)
    
data3 = {
    'x3': x,
    'J3': J3
}

df_J3 = pd.DataFrame(data3)   
print(df_J3)
print(J2)
```

```python
L=[]
iteration=0
for l in x:
   print("_______________________________________________________________")
   print(f"At x2 = {l}")
   for k in u:
       l2=-1*(l)+1+k
       if l2>2:
           l2=2
           L.append(int(l2))
       elif l2<-2:
           l2=-2
           L.append(int(l2))
       else:
           L.append(int(l2)) 
    
   print(f"x3 set for J3: {L}")
   
   #Now we have L /(x3)=[]
   J_val=[]
   for i in range(3):
       filtered_df = df_J3[df_J3['x3'] == L[i]]
       J_val.append(int(filtered_df['J3'].values[0]))
   L=[]    
   print(f"J3_curr : {J_val}")
   expr1 = 2 * abs(l) + abs(u[0]) + J_val[0]
   expr2 = 2 * abs(l) + abs(u[1]) + J_val[1]
   expr3 = 2 * abs(l) + abs(u[2]) + J_val[2]
   j_expressions=[expr1, expr2, expr3]
   J_select=min(j_expressions)
   u_curr_ind=j_expressions.index(J_select)
   u_curr=u[u_curr_ind]
   u2[iteration]=u_curr
   print(f"J* for x2: {J_select}")
   print(f"Corresponding u for min J_select: {u_curr}")
   J2[iteration]=J_select
   iteration+=1
   print(J2)

print("_______________________________________________________________") 
print(J2)    
print(u2)       
```

c) Assume now that the constant term 1 in the previous dynamics can sometimes be 0 with probability 0.4. We can now write the dynamics as

$$
x_{n+1} = 
\begin{cases} 
    -x_n + \omega_n + u_n & \text{if } -2 \leq -x_n + \omega_n + u_n \leq 2 \\
    2 & \text{if } -x_n + \omega_n + u_n > 2 \\
    -2 & \text{else}
\end{cases}
$$

where $ x_n \in \{-2, -1, 0, 1, 2\} $, $ u_n \in \{-1, 0, 1\} $ and $ \omega_n \in \{0, 1\} $ is a random variable with probability distribution $ p(\omega_n = 0) = 0.4 $, $ p(\omega_n = 1) = 0.6 $. The cost function to minimize becomes

$$
J = E \left( \sum_{k=0}^2 \left( 2|x_k| + |u_k| + x_3^2 \right) \right)
$$

Use the dynamic programming algorithm to solve the finite horizon optimal control problem that minimizes \( J \). Show the different steps of the algorithm and present the results in a table including the cost-to-go and the optimal control at every stage.


```python
u=np.array([-1, 0, 1])
x=np.array([-2,-1,0,1,2])

u2_prob=np.zeros(5)
u1_prob=np.zeros(5)
u0_prob=np.zeros(5)
J2_prob=np.zeros(5)
J1_prob=np.zeros(5)
J0_prob=np.zeros(5)

J3=[]
for i in x:
    i*=i
    J3.append(i)
    
data3_prob = {
    'x3': x,
    'J3': J3
}

df_J3_prob = pd.DataFrame(data3_prob)   
print(df_J3_prob)
```

### a. Handling of Uncertainty and Cost Calculations

The **deterministic model** assumes full predictability in the system's evolution, leading to decisions that may incur higher costs by not adapting to potential variations. In contrast, the **probabilistic model** incorporates uncertainty by considering multiple possible outcomes with associated probabilities, optimizing for expected costs and potentially resulting in lower average costs. By leveraging probability distributions, the probabilistic model balances decisions across likely scenarios, which can produce a more efficient control strategy than the deterministic approach that may overcompensate for worst-case scenarios.

### b. Optimal Control Strategy and Consistency

In the **probabilistic model**, adjustments in controls—such as shifting $u_0$ from $0.0$ to $-1.0$ at $x_0 = -2$—show an adaptive approach that minimizes expected costs by anticipating possible variations. However, for most states, both models agree on the optimal control, suggesting that under certain conditions, the presence of uncertainty does not significantly impact the decision-making process, leading to consistency in control strategies.
